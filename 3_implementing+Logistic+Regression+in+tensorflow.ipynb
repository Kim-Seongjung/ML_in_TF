```python
import tensorflow as tf 
import numpy as np
```

```python
n_point=500
```

```python
x_point=[]
y_point=[]

w=0.2
b=0.3
```

```python
for i in range(n_point):
    x=np.random.normal(0.0,1.0)
    y=w*x+b+np.random.normal(0.0 ,0.3)
    x_point.append(x)
    y_point.append(y)
```

```python
import matplotlib.pyplot as plt
plt.plot(x_point , y_point , 'o' , label='Input Data')
plt.legend()
plt.show()
```

```python
import tensorflow as tf 
```

```python
W= tf.Variable(tf.random_uniform([1] , -1.0 , 1.0))
```

```python
b= tf.Variable(tf.zeros([1]))
```

```python
y=W*x_point+b
```

```python
print type(x_point)
```

```python
cost_func= tf.reduce_mean(tf.square(y-y_point))
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.1)
```

```python
train=optimizer.minimize(cost_func)
```

```python
model = tf.initialize_all_variables()
```

```python
sess=tf.Session()
```

```python
sess.run(model)
```

```python
for step in range(20):
    plt.plot(x_point , y_point , 'o' , label ='step ={0}'.format(step))
    plt.plot(x_point , sess.run(W)*x_point +sess.run(b))
    
    plt.legend()
    plt.show()
    sess.run(train)
    
```
